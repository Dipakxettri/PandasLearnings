```python
# Data Frames :

data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Alice"],
    "age": [25, 30, 35, None, 25],
    "city": ["New York", "Paris", "Paris", "London", "New York"]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.tail()
```

```python
# Drop columns
df.drop("age", axis = 1)
```

```python
#check shape
df.shape
```

```python
# df.dtype()
df.info() 
```

```python
df.describe()
```

```python
# Broadcasting 

df['age'] = df['age'] + 20
df
```

```python
# Renaming columns 

df.rename(columns = {"salert": 'salery'}, inplace = True)
df
```

```python
# unique values

df['salery'].unique()
```

```python
# count values

df['city'].value_counts()
```

```python
# creating new column
df["Promoted Salery"] = df["salery"] + 2000
df
```

```python
# ---- Data cleaning : ---- 

df.isnull().sum() # check number of null values
```

```python
df.dropna() # this drop all null values note: without any papametor
df.dropna(how = "any") # any row that had any null value
df.dropna(how = "all") # if all rows has null value then drop otherwise not
```

```python
df.fillna(0) # fill null value with 0
df['salery'].fillna(df['salery'].mean()) # filling null salery value into mean
df['city'].fillna("Uknown") # replace all null values of citu into unkown
```

```python
df["salery"].fillna(df["salery"].median()) # more mathmatically correct
```

```python
# forward fill and backward fill

df["salery"].ffill()
df["salery"].bfill()
# but these will gave errors or problums if first or last value is null 
```

```python
# replace any specific values in an row
df["name"] = df["name"].replace("Alice","Deepak")
```

```python
# Dealing with duplicate values
# df_dup = df[df.duplicated()]
df_dup = df[df.duplicated(keep="last")]#it checks whether the entire row is a duplicate of a previous row.
df_dup
```

```python
# Drop duplicate record
df = df.drop_duplicates()
df
```

```python
# Invalid values:
# lambda 
df['age'] = df['age'].apply(lambda x : x/10 if x > 20 else x)
df

```

```python
# String functions :

# df.drop("name", axis=1, inplace=True)
names = ["Deepak_Ghimire", "Shyam_Aryal", "John_Kyo", "HSH_dd"]

# creating new column adding values in it and reordering our columns
df["name"] = names
df = df[df.columns.sort_values()]
df = df[["name", "age", "city"]]

# split :
df[["first_name","last_name"]] = df["name"].str.split("_", expand=True)
df
df = df[["first_name","last_name", "age", "city"]]
df
```

```python
# change data types:
df["age"] = df["age"].fillna(0).astype(int)
```

```python
# Joins and merge :

# Joins -> left, right, outer, inner
# merge -> 

student1 = {
    "student_id": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "age": [20, 21, 19, 22, 20],
    "grade": ["A", "B", "A", "C", "B"]
}

dff = pd.DataFrame(student1)
dff

students2 = {
    "student_id": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "age": [20, 21, 19, 22, 20]
}

dfff = pd.DataFrame(students2)
dfff
```
