```python
import pandas as pd
```

```python
# creating data frame
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "age": [25, 30, 35, None, 28],
    "city": ["New York", "Paris", "Paris", "London", None]
}

df = pd.DataFrame(data)
df


```

```python
# saving our dataframe (df) as a csv file
df.to_csv("practiceData.csv", index=False)
```

```python
# exploring data
df.info()
df.head()
df.describe()
df.isna().sum()
```

```python
# Cleaning data

# fill messing ages with mean
df['age'] = df["age"].fillna(df["age"].mean()) # Finds mean and store it to updated dataframe 
df
```

```python
# Fill missing city with "Unknown"
df["city"] = df["city"].fillna("Unknown")
df
```

```python
# filter age greater then 30
df[df['age'] > 30]
```

```python
# sort by age
# df.sort_values(by='age')
df.sort_values('age')
```

```python
# Add new column
df["age_category"] = df["age"].apply(lambda x : "Young" if x < 30 else "Adult")
df
```
