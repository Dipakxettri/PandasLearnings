```python
import pandas as pd
```

```python
# creating data frame
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "age": [25, 30, 35, None, 28],
    "city": ["New York", "Paris", "Paris", "London", None]
}

df = pd.DataFrame(data)
df
```

```python
# saving our dataframe (df) as a csv file
df.to_csv("practiceData.csv", index=False)
```

```python
# exploring data
df.info()
df.head()
df.describe()
df.isna().sum()
```

```python
# Cleaning data

# fill messing ages with mean
df['age'] = df["age"].fillna(df["age"].mean()) # Finds mean and store it to updated dataframe 
df
```

```python
# Fill missing city with "Unknown"
df["city"] = df["city"].fillna("Unknown")
df
```

```python
# filter age greater then 30
df[df['age'] > 30]
```

```python
# new data set
data = {
    "ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Name": ["Alice", "Bob", "CHARLIE", "Dana", "", "Frank", "Grace", "Henry"],
    "Age": [29, None, 35, 28, 22, "thirty", 27, 40],
    "Email": [
        "alice@example.com",
        "bob_at_example.com",
        "charlie@example",
        "dana@example.com",
        "emily@example.com",
        "frank@example.com",
        "grace@example.com",
        None
    ],
    "Join_Date": [
        "2021-03-15",
        "2020/12/01",
        "March 5, 2019",
        "2019-07-21",
        "2022-01-10",
        "2021-11-30",
        "15-08-2020",
        "2018-06-01"
    ],
    "Salary": [55000, 48000, 62000, None, 50000, 58000, 51000, 70000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# lets first fill missing values :
df["Age"] = df["Age"].fillna(0)
df= df.replace("", pd.NA)
```

```python
# lets fill names
df["Name"] = df["Name"].fillna("Uknown")
```

```python
# Convert to numeric, invalid parsing becomes NaN - age

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
# fixing emails

def fixEmails(email):
    if pd.isna(email):
        return email
    if not email.endswith('@example.com'):
        return email + "@example.com"
    return email

df["Email"] = df["Email"].apply(fixEmails)
df
```
