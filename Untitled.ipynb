```python
import numpy as np
import matplotlib.pyplot as plt

# Create blank image (black background)
h, w = 300, 300
img = np.zeros((h, w, 3), dtype=np.uint8)

# White color
white = np.array([255, 255, 255], dtype=np.uint8)

# ---- CUSTOM ORIGINAL LOGO DESIGN ----
# Draw a shield outline (polygon)
shield_points = [
    (50, 40), (250, 40),
    (280, 120), (150, 260),
    (20, 120)
]

# Simple polygon fill (manual rasterization)
for y in range(h):
    for x in range(w):
        # Ray casting method
        inside = False
        j = len(shield_points) - 1
        for i in range(len(shield_points)):
            xi, yi = shield_points[i]
            xj, yj = shield_points[j]
            if ((yi > y) != (yj > y)) and \
                (x < (xj - xi) * (y - yi) / (yj - yi + 1e-9) + xi):
                inside = not inside
            j = i
        if inside:
            img[y, x] = white

# Add simple white symbol inside (circle)
cy, cx = 150, 150
radius = 40
for y in range(h):
    for x in range(w):
        if (x - cx)**2 + (y - cy)**2 < radius**2:
            img[y, x] = white

# Save numpy file
np.save("custom_logo.npy", img)

# Show the result
plt.imshow(img)
plt.axis("off")
plt.show()

```
